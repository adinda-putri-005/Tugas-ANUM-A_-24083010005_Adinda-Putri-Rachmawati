```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Interval
x = np.linspace(0, 10, 500)
```

```python
# Deret fungsi sinusoidal
f = np.sin(x) + 0.5*np.sin(2*x) + (1/3)*np.sin(3*x) + (1/4)*np.sin(4*x) + (1/5)*np.sin(5*x)
```

```python
# Hitung integral dan diferensial dari fungsi ‘f’ untuk interval H

# Diferensial (turunan) secara numerik dengan metode beda hingga
dx = x[1] - x[0]
dfdx = np.zeros_like(f)
dfdx[1:-1] = (f[2:] - f[:-2]) / (2 * dx)      # central difference
dfdx[0] = (f[1] - f[0]) / dx                  # forward di titik awal
dfdx[-1] = (f[-1] - f[-2]) / dx
# Integral (kumulatif) dengan metode trapezoid
F = np.zeros_like(f)
for i in range(1, len(x)):
    F[i] = F[i-1] + 0.5 * (f[i] + f[i-1]) * (x[i] - x[i-1])

# atau integral total:
integral_total = np.trapz(f, x)

print(f"dx = {dx}")
print(f"Integral total ∫ f dx = {integral_total}")
```

```python
# Plot hasil
plt.figure(figsize=(10, 6))

plt.subplot(3, 1, 1)
plt.plot(x, f, label='f(x)')
plt.title('Fungsi f(x)')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(x, dfdx, label="f'(x)", color='red')
plt.title('Turunan f(x)')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(x, F, label='∫ f dx', color='yellow')
plt.title('Integral kumulatif f(x)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Hitung secara terpisah bagian integral di atas dan di bawah y=0

# Integral total
integral_total = np.trapz(f, x)

# Integral di atas (f>0) & di bawah (f<0) sumbu x
f_pos = np.where(f > 0, f, 0)   # sisakan hanya bagian positif
f_neg = np.where(f < 0, f, 0)   # sisakan hanya bagian negatif

integral_above = np.trapz(f_pos, x)     # luas di atas y=0
integral_below = np.trapz(f_neg, x)     # nilai negatif (luas di bawah sumbu x)
area_below = -integral_below            # magnitudo luas di bawah

print(f"Integral total     = {integral_total:.6f}")
print(f"Luas di atas y=0   = {integral_above:.6f}")
print(f"Luas di bawah y=0  = {area_below:.6f} (nilai integral = {integral_below:.6f})")
```

```python
# Plot fungsi dan shading area
plt.figure(figsize=(9,4))
plt.plot(x, f, 'k', label='f(x)')
plt.axhline(0, color='red', linestyle='--')
```

```python
# Shading
plt.fill_between(x, f, 0, where=f>0, color='olive', alpha=0.3, label='Area di atas')
plt.fill_between(x, f, 0, where=f<0, color='red', alpha=0.3, label='Area di bawah')

plt.title("Fungsi f(x) dengan Area Positif & Negatif")
plt.xlabel("x"); plt.ylabel("f(x)")
plt.legend(); plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot semua fungsi: f, f’, F (atas dan bawah dengan area warna berbeda)

import numpy as np
import matplotlib.pyplot as plt

# Interval & fungsi f

x = np.linspace(0, 10, 500)
f = np.sin(x) + 0.5*np.sin(2*x) + (1/3)*np.sin(3*x) + (1/4)*np.sin(4*x) + (1/5)*np.sin(5*x)

# Turunan numerik (central diff)

dx = x[1] - x[0]
dfdx = np.zeros_like(f)
dfdx[1:-1] = (f[2:] - f[:-2]) / (2 * dx)
dfdx[0]  = (f[1] - f[0]) / dx
dfdx[-1] = (f[-1] - f[-2]) / dx

# Integral kumulatif (trapezoid)

F = np.zeros_like(f)
for i in range(1, len(x)):
    F[i] = F[i-1] + 0.5*(f[i] + f[i-1])*(x[i] - x[i-1])

# Integral terpisah atas / bawah

f_pos = np.where(f > 0, f, 0)
f_neg = np.where(f < 0, f, 0)
integral_above = np.trapz(f_pos, x)
integral_below = np.trapz(f_neg, x)      # nilai negatif
area_below     = -integral_below         # magnitudo

print(f"Integral total  = {np.trapz(f, x):.6f}")
print(f"Area di atas    = {integral_above:.6f}")
print(f"Area di bawah   = {area_below:.6f} (nilai integral = {integral_below:.6f})")
```

```python
# Plot semua grafik

fig, axes = plt.subplots(3, 1, figsize=(10, 9))

# Grafik f(x)
axes[0].plot(x, f, label='f(x)')
axes[0].fill_between(x, f, 0, where=f>0, color='olive', alpha=0.25)
axes[0].fill_between(x, f, 0, where=f<0, color='red',   alpha=0.25)
axes[0].axhline(0, color='gray', lw=1, ls='--')
axes[0].set_title('Fungsi f(x)')
axes[0].set_ylabel('f(x)')
axes[0].legend(); axes[0].grid(True)

# Grafik f'(x)
axes[1].plot(x, dfdx, color='purple', label="f'(x)")
axes[1].axhline(0, color='gray', lw=1, ls='--')
axes[1].set_title('Turunan f(x)')
axes[1].set_ylabel("f'(x)")
axes[1].legend(); axes[1].grid(True)

# Grafik F(x)
axes[2].plot(x, F, label='F(x) = ∫ f dx')
axes[2].fill_between(x, F, 0, where=(F >= 0), color='green',  alpha=0.25, label='F >= 0')
axes[2].fill_between(x, F, 0, where=(F <= 0), color='purple', alpha=0.25, label='F <= 0')
axes[2].axhline(0, color='gray', lw=1, ls='--')
axes[2].set_title('Integral Kumulatif F(x)')
axes[2].set_xlabel('x'); axes[2].set_ylabel('F(x)')
axes[2].legend(); axes[2].grid(True)

plt.tight_layout()
plt.show()
```
